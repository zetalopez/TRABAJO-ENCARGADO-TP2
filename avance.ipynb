```python
#IMPORTANDO DATA POR ARCHIVOS CSV DESCARGADOS

import pandas as pd
import numpy as np
data1 = pd.read_csv('bank.csv',sep = ";", header = None,skiprows=1,
                   names=['edad','trabajo','estado_civil','educacion','default','balance','casa','prestamo','contacto','dia','mes','duracion','campaña','pdia','previo','p salir','y'])
data1
```

```python
# tabla de contingencia en porcentajes relativos total con los datos descargados

dataframe1 = pd.crosstab(index=data1['estado_civil'], columns=data1['prestamo'],
            margins=True).apply(lambda r: r/len(data1) ,
                                axis=1)
dataframe1
```

```python
# IMPORTANDO DATA DESDE UN URL

import pandas as pd
import numpy as np
data2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep = ",", header = None,
                  names=['edad','tipo_de_empleo','peso_final','educacion', 
                     'años_educacion','estado _civil','ocupacion','parentesco','raza','sexo','ganancia_capital','perdida_capital','horas_semana','pais','ingresos'])
data2
```

```python
# tabla de contingencia en porcentajes relativos total
dataframe2 = pd.crosstab(index=data2['educacion'], columns=data2['ingresos'],
            margins=True).apply(lambda r: r/len(data2) ,
                                axis=1)
dataframe2
```
