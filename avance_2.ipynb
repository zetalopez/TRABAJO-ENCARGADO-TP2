```python
# IMPORTANDO DATA DESDE UN URL

import pandas as pd
import numpy as np
data2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep = ",", header = None,
                  names=['edad','tipo_de_empleo','peso_final','educacion', 
                     'años_educacion','estado _civil','ocupacion','parentesco','raza','sexo','ganancia_capital','perdida_capital','horas_semana','pais','ingresos'])
data2
```

```python
# tabla de contingencia en porcentajes relativos total
dataframe2 = pd.crosstab(index=data2['educacion'], columns=data2['ingresos'],
            margins=True).apply(lambda r: r/len(data2) ,
                                axis=1)
dataframe2
```

```python
#funciones

def obtenerMatriz(dataframe):
    np.savetxt("matriz.csv",dataframe)

    matrix = np.genfromtxt('matriz.csv')

    return matrix

def RPE(matrix):
    
    i = len(matrix)#numero de columnas del array
    j = len(matrix[0]) #numero de filas del array

    matrix2 = matrix[0:i-1,0:j-1] #se creo un nuevo array sin los ultimos elementos de cada fila , y los ultimos de cada columna, quedandonos con los restantes, siguiendo el orden del array original.
    maxc = matrix2.max(axis=0)  # obtener maximas probabilidades por columnas del array.
    maxf = matrix2.max(axis=1) # obtener maximos probabilidades por filas del array.

    sumax_columnas = maxc.sum()# suma de los elementos de maxc.

    sumax_filas = maxf.sum()# suma de los elementos de maxf.

    matrix3 = matrix[0:i-1,j-1]#obtenemos la ultima columna del array , para obtener las probabilidades marginales.

    maxtf = matrix3.max()#obtenemos el maximo del array matrix3

    matrix4 = matrix[i-1,0:j-1]#obtenemos la ultima fila del array , para obtener las probabilidades marginales.

    maxtc = matrix4.max()#obtenemos el maximo del array matrix4

    lambdas = (sumax_columnas + sumax_filas - maxtc -maxtf)/(2-maxtc-maxtf)#obteniendo el lambda simetrico

    return lambdas

    print("El estadístico requerido lambda es: ", lambdas)

    
```

```python
data = obtenerMatriz(dataframe2)
data
```

```python
RPE = RPE(data)
RPE
```
