```python
# IMPORTANDO DATA DESDE UN URL

import pandas as pd
import numpy as np
data2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',sep = ",", header = None,
                  names=['edad','tipo_de_empleo','peso_final','educacion', 
                     'a√±os_educacion','estado _civil','ocupacion','parentesco','raza','sexo','ganancia_capital','perdida_capital','horas_semana','pais','ingresos'])
data2
##Mensaje de prueba
```

```python
# tabla de contingencia en porcentajes relativos total
dataframe2 = pd.crosstab(index=data2['educacion'], columns=data2['ingresos'],
            margins=True).apply(lambda r: r/len(data2) ,
                                axis=1)
dataframe2
```

```python
#funciones

def obtenerMatriz(dataframe):
    np.savetxt("matriz.csv",dataframe)

    matrix = np.genfromtxt('matriz.csv')

    return matrix

def RPE(matrix):
    
    i = len(matrix)#numero de columnas del array
    j = len(matrix[0]) #numero de filas del array

    matrix2 = matrix[0:i-1,0:j-1] #se creo un nuevo array sin los ultimos elementos de cada fila , y los ultimos de cada columna, quedandonos con los restantes, siguiendo el orden del array original.
    maxc = matrix2.max(axis=0)  # obtener maximas probabilidades por columnas del array.
    maxf = matrix2.max(axis=1) # obtener maximos probabilidades por filas del array.

    sumax_columnas = maxc.sum()# suma de los elementos de maxc.

    sumax_filas = maxf.sum()# suma de los elementos de maxf.

    matrix3 = matrix[0:i-1,j-1]#obtenemos la ultima columna del array , para obtener las probabilidades marginales.

    maxtf = matrix3.max()#obtenemos el maximo del array matrix3

    matrix4 = matrix[i-1,0:j-1]#obtenemos la ultima fila del array , para obtener las probabilidades marginales.

    maxtc = matrix4.max()#obtenemos el maximo del array matrix4

    lambdas = (sumax_columnas + sumax_filas - maxtc -maxtf)/(2-maxtc-maxtf)#obteniendo el lambda simetrico

    return lambdas

    print("El estad√≠stico requerido lambda es: ", lambdas)

    
```

```python
#1> asumimos que una variable es predominante que otra (en este caso las columnas sobre las filas)
#llamaremos a esta funcion "ydex"
#donde a(una matriz en lista) es el prametro de entrada para esta funcion

def ydex(a):
    
    n=len(a)   #Lectura de la cantidad de elementos de la Matriz    m=len(a[0]) #Longitud de cada elemento 

    ja=a[:n-1] #ja toma el valor de la matriz anterior desde el elemento 0 al n-1 
    print(ja)

    n1=len(ja) #Lectura de la cantidad de elemntos de ja
    m1=len(ja[0]) #Longitud de cada elemento

    print(m1)
    print(n1)
    ma1=[] #Creamos una lista
    for i in range(m1):
        ma1.append([0]*(n1)) #Agregamos a la lista vacia elementos [0]
    print(ma1)



    for i in range(m1): 
        for j in range(n1):
            ma1[i][j]=ja[j][i]
    print(ma1)

    s=0
    for i in range(len(ma1)-1): #Para cada elemento de ma1
        a=max(ma1[i]) #le asignamos el valor max de cada elemento para a
        s=s+a #Sumamos los valores maximos
        a=0
    pmar1=max(ma1[-1]) #Hallamos el valor maximo del ultimo elemento
    print(s)
    print(pmar1)
    resul1=(s-pmar1)/(1-pmar1)
    print(resul1)
    return[s,pmar1,resul1] 

prueba=ydex(matriz)
print('la suma de maximas probabilidades por columnas:',prueba[0])
print('la mayor probabilidad marginal:',prueba[1])
print('el estadistico es:',prueba[2])
```

```python
#2> asumimos que una variable es predominante que otra (en este caso las filas sobre las columnas)
#llamaremos a esta funcion "xdey"
#donde a(una matriz en lista) es el prametro de entrada para esta funcion
def xdey(a):
    n=len(a)    
    m=len(a[0])

    ma2=[]
    for i in range(n):
        b=a[i]
        c=b[:m-1]
        ma2.append(c)
    print(ma2)
    
    s=0
    for i in range(len(ma2)-1):
        a=max(ma2[i])
        s=s+a
        #print(s)
    print(round(s,3))

    pmar2=max(ma2[-1])

    resul2=(s-pmar2)/(1-pmar2)
    print(resul2)
    return[s,pmar2,resul2]
prueba=xdey(matriz)
print('la suma de maximas probabilidades por filas:',prueba[0])
print('la mayor probabilidad marginal:',prueba[1])
print('el estadistico es:',prueba[2])
```

```python
# las versiones anteriores se dednominan asimetricos porque al intercambiar la variable respuestas con la explicativa, el valor de los estadisticos no siempr es igual
#pero requeriremos de simtrica del coeficinete lambda
#3> para este caso haremos una version simetrica
def asime(a):
    t=  ydex(a)
    t1=t[0]
    t2=t[1]
    p=  xdey(a)
    p1=p[0]
    p2=p[1]

    lam=(t1+p1-t2-p2)/(2-t2-p2)
    return lam
prueba=asime(matriz)
print('el estadistico lambda es:',prueba)
    
```
